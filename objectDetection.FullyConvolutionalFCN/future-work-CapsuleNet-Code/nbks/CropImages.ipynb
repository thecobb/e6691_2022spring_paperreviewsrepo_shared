# Final Code to resize, crop, and resave image data:
## Loading, preprocessing, and writing of image data


```python
def Simple_PolypCrop(img):#This function takes in an image and crops it to a preset size
   
    #resize image so that they are all the same shape  
    resized_img = cv2.resize(img, (576, 576))
    
    #Crop image to remove black background  
    cropped_Image=resized_img[30:540, 50:545]

    return cropped_Image
```
